```python
data.drop(columns=['title'],inplace=True)
```

```python
# converting into lower case
```

```python
data['text']= data['text'].str.lower()
data['text']
```

```python
#Removal of speacial character
import re
```

```python
# Function to remove special characters
def remove_special_characters(text):
    pattern = r'[^A-Za-z\s]'
    return re.sub(pattern, '', text)
```

```python
data['text']= data['text'].apply(remove_special_characters)
```

```python
data['text']
```

```python
# Function to remove Back space characters
def remove_back_characters(text):
    pattern = r'[\\\n]'
    return re.sub(pattern, '', text)
```

```python
data['text']= data['text'].apply(remove_back_characters)
data['text'].head(4)
```

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
```

```python
def remove_stopwords(text):
  stopwords_list=set(stopwords.words('english'))
  words=text.split()
  filtered_words=[word for word in words if word.lower() not in stopwords_list]
  return ' '.join(filtered_words)


```

```python
data['text']= data['text'].apply(remove_stopwords)
```

```python
data
```

```python
x= data['text']
y= data['type']
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest= train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf= TfidfVectorizer()
```

```python
xtrain_tfidf= tfidf.fit_transform(xtrain)
xtest_tfidf= tfidf.transform(xtest)
```

```python
xtrain_tfidf.shape
```

```python
xtest_tfidf.shape
```

```python
xtrain_tfidf.toarray()
```

```python
ytrain
```

```python
from sklearn.svm import SVC
```

```python
model=SVC()
```

```python
model.fit(xtrain_tfidf,ytrain)
```

```python
y_pred=model.predict(xtest_tfidf)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
accuracy= accuracy_score(ytest,y_pred)
accuracy
```
